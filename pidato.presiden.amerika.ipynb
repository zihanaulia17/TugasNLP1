```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
speech_url = 'https://www.itapuih.com/2019/07/pidato-bahasa-inggris-barack-obama.html'
```

```python
response = requests.get(speech_url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
text = soup.get_text()
```

```python
text = text.lower()
text = ''.join(c if c.isalnum() or c.isspace() else ' ' for c in text)
words = text.split()
```

```python
word_counter = Counter(words)
```

```python
most_common_words = word_counter.most_common(20)
```

```python
df = pd.DataFrame(most_common_words, columns=['Word', 'Frequency'])
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Word'], df['Frequency'])
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('20 Most Common Words in Presidential Speech')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
